```python
import pandas as pd
from datetime import datetime, timedelta
import os
```

```python
out_folder = 'outputs_bme280'
out_path = os.path.join(os.getcwd(), out_folder)

in_folder = 'inputs'
in_path = os.path.join(os.getcwd(), in_folder)
in_path
```

```python
# list files in the directory 
files = [f for f in os.listdir(in_path) if os.path.isfile(os.path.join(in_path, f))]
files
```

```python
minutes = []
hours = []
api = []

for file in files:
    file_path = os.path.join(in_path, file)
    df = pd.read_csv(file_path, index_col=None)

    if 'minutes' in file:
        minutes.append(df)
    if 'api' in file:
        api.append(df)
    if 'hours' in file:
        hours.append(df)
        
```

```python
df_minutes = pd.concat(minutes)
df_hours = pd.concat(hours)
df_api = pd.concat(api)

df_minutes.to_csv(os.path.join(out_path, 'minutes_5to9.csv'), index=False)
df_hours.to_csv(os.path.join(out_path, 'hours_5to9.csv'), index=False)
df_api.to_csv(os.path.join(out_path, 'api_5to9.csv'), index=False)
```
